### Basic Preprocessing

```python
tweet_column = 'Tweet'

# Function to clean the tweets
def clean_tweet(tweet):
    # Remove URLs
    tweet = re.sub(r'http\S+|www\S+|https\S+', '', tweet, flags=re.MULTILINE)
    # Remove user @ references and '#' from tweet
    tweet = re.sub(r'\@\w+|\#','', tweet)
    # Remove special characters, numbers, punctuations (except for the apostrophe)
    tweet = re.sub(r'[^a-zA-Z\'\s]', '', tweet)
    # Convert to lowercase
    tweet = tweet.lower()
    return tweet
```

```python
# Apply the cleaning function to the tweet column
df[tweet_column] = df[tweet_column].apply(clean_tweet)
```

```python

# Display the first few rows of the cleaned tweets
df.head()
```

```python
df = df.rename(columns = {'AverageAnnotation':'Anotation_Score'})
```

This gives the dataframe shape of (7475, 3). The "AverageAnnotation" column contains float numeric values that fall between -2.75 and 3.0. The mean AverageAnnotation score is 0.26 and the median AverageAnnotation score is 0.25. The "Tweet" and "Username" columns both contain strings.

- The length of a tweet has almost no correlation with the average annotation score. Additionally, the average length of the words in a tweet is very weakly correlated with the average annotation score.

```python
# Descriptive Statistics
mean_score = merged_df['Anotation_Score'].mean()
median_score = merged_df['Anotation_Score'].median()
mode_score = merged_df['Anotation_Score'].mode()
```

```python
# Word Frequency Analysis
all_words = ' '.join(merged_df['Tweet']).split()
word_freq = Counter(all_words)
most_common_words = word_freq.most_common(10)
```

```python
# Calculate average word length and sentence length
merged_df['average_word_length'] = merged_df['Tweet'].apply(lambda x: np.mean([len(word) for word in x.split()]))
merged_df['sentence_length'] = merged_df['Tweet'].apply(lambda x: len(x.split()))
```

```python
# Correlation Analysis
correlation_matrix = merged_df[['Anotation_Score', 'average_word_length', 'sentence_length']].corr()
```

```python
# Print the statistics
print(f"Mean Optimism Score: {mean_score}")
print(f"Median Optimism Score: {median_score}")
print(f"Mode Optimism Score: {mode_score.tolist()}")
print("\nMost Common Words:")
print(most_common_words)
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

The correlation matrix shows there is not a linear relationship between annotation score and the average word length of tweets or the sentence length

```python
# N-Gram Analysis for Bigrams and Trigrams
def n_gram_analysis(dataframe, n=None):
    count_vectorizer = CountVectorizer(ngram_range=(n, n), stop_words='english')
    ngrams = count_vectorizer.fit_transform(dataframe['Tweet'])
    count_values = ngrams.toarray().sum(axis=0)
    vocab = count_vectorizer.vocabulary_
    df_ngram = pd.DataFrame(sorted([(count_values[i],k) for k,i in vocab.items()], reverse=True)
                             ).rename(columns={0: 'frequency', 1:'n-gram'})
    print(df_ngram.head())

# User Behavior Analysis
def user_behavior_analysis(dataframe):
    dataframe['user_activity'] = dataframe['ProfileInfo'].apply(lambda x: 'high' if 'active' in x else 'low')
    sns.boxplot(data=dataframe, x='user_activity', y='Anotation_Score')
    plt.title('Distribution of Annotation Scores of Users by Activity')

    plt.show()
```

```python
n_gram_analysis(merged_df, n=2)  # For bigrams
```

```python
n_gram_analysis(merged_df, n=3)  # For trigrams
```

The box and whisker plot of annotation scores by user activity shows that users that are highly active tend post tweets that are pessimistic. Users that are less active tend to post tweets that are pretty evently optimistic and pessimistic (there is slightly more volume of optimistic tweets).
